```python
import stanza
import os
from stanza.server import CoreNLPClient
import pandas as pd
import numpy as np
```

```python
stanza.download('en') # download English model
nlp = stanza.Pipeline('en') # initialize English neural pipeline
```

```python
# Download the Stanford CoreNLP package with Stanza's installation command
# This'll take several minutes, depending on the network speed
corenlp_dir = './corenlp'
stanza.install_corenlp(dir=corenlp_dir)

# Set the CORENLP_HOME environment variable to point to the installation location
os.environ["CORENLP_HOME"] = corenlp_dir
```

```python
# Examine the CoreNLP installation folder to make sure the installation is successful
!ls $CORENLP_HOME
```

```python
# Construct a CoreNLPClient with some basic annotators, a memory allocation of 4GB, and port number 9001
client = CoreNLPClient(timeout=150000000,
    annotators=['openie'],
    memory='4G',
    endpoint='http://localhost:9002',
    be_quiet=True)
client.start()
```

```python
text = "Albert Einstein was a German-born theoretical physicist. He developed the theory of relativity."
document = client.annotate(text, output_format='json')
triples = []
for sentence in document['sentences']:
    for triple in sentence['openie']:
        triples.append({
           'subject': triple['subject'],
           'relation': triple['relation'],
            'object': triple['object']
        })
print(triples)
```

```python
df = pd.read_csv('/content/drive/MyDrive/Tesi & Stage/Code/PUBHEALTH/train.tsv', sep='\t', nrows = 500) #nrows = 500 per provare
df.head()
```

```python
# Scarica e inizializza il modello e il server CoreNLP una sola volta
stanza.download('en')
nlp_model = stanza.Pipeline('en', package='genia')
corenlp_dir = './corenlp'
stanza.install_corenlp(dir=corenlp_dir)
os.environ["CORENLP_HOME"] = corenlp_dir

# Construct a CoreNLPClient with some basic annotators, a memory allocation of 4GB, and port number 9001
client = CoreNLPClient(
    annotators=['openie'],
    memory='4G',
    endpoint='http://localhost:9006',
    be_quiet=True)

# Definisci la funzione extract_openie_triplets utilizzando le variabili globali
def extract_openie_triplets(text):
    # Utilizza il modello e il server CoreNLP globali
    document = client.annotate(text, output_format='json')

    # Estrai le triplette OpenIE dal documento annotato
    triples = []
    for sentence in document['sentences']:
      for triple in sentence['openie']:
        triples.append({
           'subject': triple['subject'],
           'relation': triple['relation'],
            'object': triple['object']
        })
    print(triples)
```

```python
for i in df['claim']:
  triple = extract_openie_triplets(i)
  print(i)
```
